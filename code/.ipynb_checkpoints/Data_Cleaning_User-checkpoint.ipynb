```python
#importing important libraries
import pandas as pd
import numpy as np
import json
```

```python
#since pd.readjson will cause memeory error, we read the file line by line
path = '../yelp_academic_dataset_user.json'
user = []
with open(path, encoding = 'utf-8') as fin:
    for line in fin:
        line_contents = json.loads(line)
        user.append(line_contents)
```

```python
user = pd.DataFrame(user)
user.shape
```

```python
user.head()
```

```python
user.to_csv('../data/output_csv/user.csv',index = False)
```

```python
review = pd.read_csv('../data/output_csv/review_small.csv')
```

```python
review.head()
```

```python
#filter the users from PA
unique_PA_users = review.user_id.unique()
```

```python
user_index = []
for i in range(user.shape[0]):
    if user.user_id[i] in unique_PA_users:
        user_index.append(i)
```

```python
user_df = user.iloc[user_index,:]
```

```python
#fix the index
user_df.index = range(user_df.shape[0])
user_df.head()
```

```python
#Check missing values
user_df.isnull().sum()
```

```python
# cleaning elites
user_df['is_elite'] = ~np.array(user_df.elite.isnull())
user_df.is_elite.describe()
```

```python
# too many values, may be just take year.
len(user_df.yelping_since.unique()) 
```

```python
# get the year
yr_ls = [eval(x.split('-')[0]) for x in user_df.yelping_since]
user_df['year'] = yr_ls
user_df
```

```python
user_df.to_csv('../data/output_csv/user_PA.csv',index = False)
```
