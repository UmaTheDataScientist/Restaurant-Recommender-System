```python
import pandas as pd
import numpy as np
import json
```

```python
#since pd.readjson will cause memeory error, we read the file line by line
path = 'data/yelp_academic_dataset_review.json'
review = []
with open(path, encoding = 'utf-8') as fin:
    #i = 0 
    for line in fin:
        line_contents = json.loads(line)
        review.append(line_contents)       
```

```python
review = pd.DataFrame(review)
review.shape
```

```python
review.head()
```

```python
review.drop(['review_id','useful', 'funny','cool','date'], axis=1,inplace = True)
```

```python
review.shape
```

```python
review.head()
```
